```python
import requests
import time

API_URL = "https://api-inference.huggingface.co/models/BaranKanat/BerTurk-SpamSMS"
headers = {"Authorization": "Bearer hf_YOURAPI"}

#payload = {"inputs": "2000 TL DENEME BONUSU KAZANDINIZ !!! YATIRIM SARTI YOK KAZANC ve CEKIM LIMITI YOK."}
payload = {"inputs": "Merhaba Ahmet, nasılsın?"}


# Try to get the response and handle potential errors
try:
    response = requests.post(API_URL, headers=headers, json=payload)
    response.raise_for_status()  # Raise an exception for bad status codes (like 503)
    result = response.json()
    print(result)
except requests.exceptions.RequestException as e:
    print(f"Error: {e}")
    print("Retrying in 5 seconds...")
    time.sleep(5)  # Wait for 5 seconds before retrying
    # You can add a loop here to retry multiple times if needed
    # ...
```
